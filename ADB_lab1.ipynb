```python
import numpy as np

a = np.sqrt(2)

#creating matrix M
M = np.array([[a,1,-a],[0,1,1],[-a,a,1]])
print("Matrix M:\n {0}".format(M))

#inverting matrix M
Minv = np.linalg.inv(M)
print("Matrix M inverted:\n {0}". format(Minv))

#transposing array M
Mt = np.transpose(M)
print("Matrix M transposed:\n {0}". format(Mt))

#calculting determinant of the matrix M
Mdet = np.linalg.det(M)
print("detM = {0}".format(Mdet))
```

```python
print("M 1x1 = {0}".format(M[0,0]))
print("M 3x3 = {0}".format(M[2,2]))
print("M 3x2 = {0}".format(M[2,1]))

w1 = M[2,:]
w2 = M[1,:]
print("vector w1 = {0}".format(w1))
print("vector w2 = {0}".format(w2))
```

```python
import math
import matplotlib.pyplot as plot
import numpy as np

def Q(m,v):
    return (m*v**2)/2
#changing "m" from g to kg and "v" from km/h to m/s so that result is given in Joules 
m = 2500*math.pow(10,-3)*(1000/3600)**2
v = 60

q1 = Q(m,v)

#1J = 0,238846cal = 0,000238846kcal

q1c = q1*0.000238846

print('q1 = {0} J = {1} kcal'.format(q1,q1c))

m = 3000*math.pow(10,-3)*(1000/3600)**2
v = 200 - np.linspace(0,200)
q2 = Q(m,v)

plot.plot(v,q2[0] - q2)
plot.title('change in heat')
plot.xlabel('v[km/h]')
plot.ylabel('Q[J]')
plot.legend(['Q(v)'])
plot.show()

plot.semilogx(v,q2[0] - q2)
plot.title('change in heat (logarithmic scale)')
plot.xlabel('v[km/h]')
plot.ylabel('Q[J]')
plot.legend(['Q(v)'])
plot.show()
```

```python
class Quaternion:

    def __init__(self, a, b, c, d):
        self.__a = a
        self.__b = b
        self.__c = c
        self.__d = d

    def __str__(self):
        return 'Quaternion = {0} + {1}i + {2}j + {3}k'.format(self.__a, self.__b, self.__c, self.__d)
        
    def __add__(self, another_quaternion):
        return Quaternion(self.__a + another_quaternion.__a, self.__b + another_quaternion.__b , self.__c + another_quaternion.__c, self.__d + another_quaternion.__d)
    
    def __sub__(self, another_quaternion):
        return Quaternion(self.__a - another_quaternion.__a, self.__b - another_quaternion.__b , self.__c - another_quaternion.__c, self.__d - another_quaternion.__d)
    
    def __mul__(self, another_q):
        return Quaternion(self.__a * another_q.__a - self.__b * another_q.__b - self.__c * another_q.__c - self.__d * another_q.__d, self.__b * another_q.__a + self.__a * another_q.__b + self.__c * another_q.__d - self.__d * another_q.__c, self.__a * another_q.__c - self.__b * another_q.__d + self.__c * another_q.__a + self.__d * another_q.__b, self.__a * another_q.__d + self.__b * another_q.__c - self.__c * another_q.__b + self.__d * another_q.__a)
 
    def __floordiv__(self, another_q):
        return Quaternion(self.__a * another_q.__a - self.__b * another_q.__b * -1 - self.__c * another_q.__c * -1 - self.__d * another_q.__d * -1, self.__b * another_q.__a + self.__a * another_q.__b * -1 + self.__c * another_q.__d * -1 - self.__d * another_q.__c * -1, self.__a * another_q.__c * -1 - self.__b * another_q.__d * -1 + self.__c * another_q.__a + self.__d * another_q.__b * -1, self.__a * another_q.__d * -1 + self.__b * another_q.__c * -1 - self.__c * another_q.__b * -1 + self.__d * another_q.__a)

    def conj(self):
        self.__a = self.__a * -1
        self.__b = self.__b * -1
        self.__c = self.__c * -1
        self.__d = self.__d * -1
        
    def norm(self):
        n = np.sqrt(self.__a**2 + self.__b**2 + self.__c**2 +self.__d**2)
        self.__a = self.__a / n
        self.__b = self.__b / n
        self.__c = self.__c / n
        self.__d = self.__d / n
                          
q1 = Quaternion(1,2,-3,4)
print('q1')
print(q1)

q2 = Quaternion(4,3,1,7)
print('q2')
print(q2)

q3 = q1 + q2
print('q3 = q1 + q2')
print(q3)

q4 = q2 - q1
print('q4 = q2 - q1')
print(q4)

q5 = q1 * q2
print('q5 = q1 * q2')
print(q5)

q6 = q1 // q2
print('q6 = q1 / q2')
print(q6)

q1.conj()
print('conj(q1)')
print(q1)

q1.norm()
print('norm(q1)')
print(q1)
    
```
