```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt

Normal_dict = {
    "Normal":np.random.normal(size=10000),
    "Normal_M=2_SD_=3":np.random.normal(loc=2.,scale=3.,size=10000),
    "Normal_M=4_SD=2":np.random.normal(loc=4.,scale=2.,size=10000)
}
#generating normal distribution data
```

```python
Dataframe = pd.DataFrame.from_dict(Normal_dict)
#saving normal_dict as pandas to use pands tools
```

```python
MeltedData = Dataframe.melt()
Graph = sb.FacetGrid(MeltedData, col="variable", sharex=True, col_wrap = 3)
Graph.map(plt.hist, "value", cumulative=False, bins=128)
#depending on mean and standard deviation, distribution of values differs
```

```python
GraphCum = sb.FacetGrid(MeltedData, col="variable", sharex=True, col_wrap=3)
GraphCum.map(plt.hist, "value", cumulative=True, bins=64)
#cumulative distribution
```

```python
#expit function is inverse logit function
from scipy.special import expit
ExpitData = expit(Dataframe)
MeltedExpit = ExpitData.melt()
GraphExpit = sb.FacetGrid(MeltedExpit, col="variable", sharex=True, col_wrap=3)
GraphExpit.map(plt.hist, "value", cumulative=False, bins=64)

GraphExpitCum = sb.FacetGrid(MeltedExpit, col="variable", sharex=True, col_wrap=3)
GraphExpitCum.map(plt.hist, "value", cumulative=True, bins=64)
```

```python
Poisson_dict = {
    "PoissonInterval=1":np.random.poisson(1,100000),
    "PoissonInterval=4":np.random.poisson(4,100000),
    "PoissonInterval=9":np.random.poisson(9,100000),
    "PoissonInterval=16":np.random.poisson(16,100000)
}

Dataframe_Poisson = pd.DataFrame.from_dict(Poisson_dict)
Melted_Poisson = Dataframe_Poisson.melt()

Graph_Poisson = sb.FacetGrid(Melted_Poisson, col="variable", sharex=True, col_wrap=4)
Graph_Poisson.map(plt.hist, "value", cumulative=False, bins=64)
Graph_Poisson_Cum = sb.FacetGrid(Melted_Poisson, col="variable", sharex=True, col_wrap=4)
Graph_Poisson_Cum.map(plt.hist, "value", cumulative=True, bins=128)

#poisson distribution is defines by sqrt(lambda), where lambda is expected value
```

```python
Beta_dict = {
    "Beta_alpha=1_beta=1":np.random.beta(1,1,10000),
    "Beta_alpha=5_beta=1":np.random.beta(5,1,10000),
    "Beta_alpha=1_beta=5":np.random.beta(1,5,10000),
    "Beta_alpha=5_beta=5":np.random.beta(5,5,10000),
    "Beta_alpha=3_beta=27":np.random.beta(3,27,10000),
    "Beta_alpha=8_beta=2":np.random.beta(8,2,10000)
}

Dataframe_Beta = pd.DataFrame.from_dict(Beta_dict)
Melted_Beta = Dataframe_Beta.melt()
Graph_Beta = sb.FacetGrid(Melted_Beta, col="variable", sharex=True, col_wrap=3)
Graph_Beta.map(plt.hist, "value", cumulative=False, bins=64)
Graph_Beta_Cum = sb.FacetGrid(Melted_Beta, col="variable", sharex=True, col_wrap=3)
Graph_Beta_Cum.map(plt.hist, "value", cumulative=True, bins=128)

#expected value is alpha/(alpha+beta), where alpha is number of successes and beta is number of failures
```

```python
Lognormal_dict = {
    "Lognormal_M=0_S=1":np.random.lognormal(size=10000),
    "Lognormal_M=2_S=0.1":np.random.lognormal(2,0.1,10000),
    "Lognormal_M=1_S=0.5":np.random.lognormal(1,0.5,10000),
    "Lognormal_M=5_S=1":np.random.lognormal(5,1,10000),
    "Lognormal_M=0_S=0.1":np.random.lognormal(0,0.1,10000),
    "Lognormal_M=0.1_S=1":np.random.lognormal(0.1,1,10000),
}

Dataframe_Lognormal = pd.DataFrame.from_dict(Lognormal_dict)
Melted_Lognormal = Dataframe_Lognormal.melt()

Graph_Lognormal = sb.FacetGrid(Melted_Lognormal, col="variable", sharex=False, sharey=False, col_wrap=3)
Graph_Lognormal.map(plt.hist, "value", cumulative=False, bins=64)
Graph_Lognormal_Cum = sb.FacetGrid(Melted_Lognormal, col="variable", sharex=False, sharey=False, col_wrap=3)
Graph_Lognormal_Cum.map(plt.hist, "value", cumulative=True, bins=128)

#the lognormal distribution is determined by mean and random variable values are under logarithm
#the whole expression needs to be divided by x
```
