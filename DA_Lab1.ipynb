```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv(filepath_or_buffer="https://raw.githubusercontent.com/jgajoch/public_lectures/master/Data%20Analytics/Lab%201%20-%20Python%20review/Data1.csv")
#data is uploaded from github repository
```

```python
data
```

```python
data = data.set_index(data.columns[0])
data
#data's index changed to it's first column - dates
```

```python
kol = ['theta_1', 'theta_2', 'theta_3', 'theta_4', 'theta_5', 'theta_6']
data[kol].plot(subplots=True, figsize=(20,15))
plt.show()
#data will be presented on 6 subplots, one for each theta
```

```python
data2 = data.melt()
data2
#melting the data will result in reshaping data table so variables in columns are assigned to their values
#dataIndex is lost in that process
```

```python
graph = sb.FacetGrid(data2, col="variable", sharex=False, col_wrap=3)
graph.map(plt.hist,"value")
#now with data melted it's possible to create histograms which are usefull for presenting probability
```

```python
for i in kol:
    plt.figure()
    sb.distplot(data[i], kde=True, hist=False)
```

```python
data2018 = data['2018-01-01':'2019-01-01']

kol2018 = data.columns
kol2018 = kol2018[0:4]
print(kol2018)
reduced2018 = data2018[kol2018]

melted2018 = reduced2018.melt()

reduced2018.plot(subplots=True,figsize=(20,15))
plt.show()

#repeated analisis of columns theta_1 to theta_4 for year 2018
#dat presented in 4 subplots
```

```python
plt.figure()
grid = sb.FacetGrid(melted2018, col="variable")
grid.map(plt.hist, "value")
#data presented in histograms
```

```python
for i in kol2018:
    plt.figure()
    sb.distplot(reduced2018[i], kde=True, hist=False)
#results show plots similiar to gaussian distribution
```
